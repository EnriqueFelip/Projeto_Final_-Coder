```python
import requests
import pandas as pd
from datetime import datetime
from plyer import notification

# Função para enviar notificações
def alerta(nivel, base, etapa, erro=""):
    '''
        Alerta de falha de carregamento de base de dados
    '''
    now = str(datetime.now())
    msg = f"Falha no carregamento da base {base} na etapa {etapa}.\n{now}\nErro: {erro}"

    if nivel == 1:
        title = 'ATENÇÃO: Alerta Baixo'
    elif nivel == 2:
        title = 'ATENÇÃO: Alerta Médio'
    elif nivel == 3:
        title = 'ATENÇÃO: Alerta Alto'
    else:
        print("Nível", nivel, "não disponível!")
        return

    notification.notify(
        title=title,
        message=msg,
        app_name='alerta',
        timeout=10
    )

# Função para fazer a requisição e extrair dados
def extrair_dados(url, base_name):
    try:
        resposta = requests.get(url)
        resposta.raise_for_status()  # Levanta um erro se o status code não for 200
        dados = resposta.json()  # Supondo que a resposta seja em JSON
        print(f"Dados de {base_name} carregados com sucesso")
        return dados
    except requests.exceptions.HTTPError as http_err:
        if resposta.status_code == 404:
            alerta(nivel=1, base=base_name, etapa="Requisição HTTP", erro=str(http_err))
        elif resposta.status_code == 500:
            alerta(nivel=3, base=base_name, etapa="Requisição HTTP", erro=str(http_err))
        else:
            alerta(nivel=2, base=base_name, etapa="Requisição HTTP", erro=str(http_err))
    except Exception as err:
        alerta(nivel=3, base=base_name, etapa="Requisição HTTP", erro=str(err))


# URLs a serem extraídas
urls = {
    "IBGE": "https://brasilapi.com.br/api/ibge/uf/v1",
    "cidade": "https://brasilapi.com.br/api/cptec/v1/cidade",
    "Feriados": "https://brasilapi.com.br/api/feriados/v1/2024"  # Exemplo de ano fixo
}


# Função main para encapsular a lógica do script
def main():
    # URLs a serem extraídas
    urls = {
        "IBGE": "https://brasilapi.com.br/api/ibge/uf/v1",
        "cidade": "https://brasilapi.com.br/api/cptec/v1/cidade",
        "Feriados": "https://brasilapi.com.br/api/feriados/v1/2024"  # Exemplo de ano fixo
    }

    # Dicionário para armazenar os dados extraídos
    dados_extraidos = {}

    # Extrair dados de cada URL
    for base_name, url in urls.items():
        dados = extrair_dados(url, base_name)
        if dados:
            dados_extraidos[base_name] = dados

    # Visualizar os dados em forma de tabela
    for base_name, dados in dados_extraidos.items():
        df = pd.DataFrame(dados)
        print(f"Tabela de {base_name}:")
        print(df)
        print("\n")

# Executar a função main
if __name__ == "__main__":
    main()
```
